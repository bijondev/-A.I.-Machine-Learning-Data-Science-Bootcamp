# Matplotlib example workflow

```python
# 1. prepare data
x=[1,2,3,4]
y= [11,22,33,44]

#2. setup plot
fig, ax=plt.subplots(figsize=(10,5))

# 3. plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis"
       )

# 5. Save & show (you save the whole figure)
fig.savefig("/content/drive/MyDrive/data_science/matplotlib/sample-plot.png")

```

# Which one should you use?
* pyplot
* matplotlib oo method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter", x="age", y="chol", c="target") # pyplot method
```

```python
# 00 method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax)

# ax.set_xlim([45, 100])
```

```python
fix, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

ax.set(title="Heart Disease and Cholesterol levels", xlabel="Age", ylabel="Cholesterol")
ax.legend(*scatter.legend_elements(), title="Target");

ax.axhline(over_50["chol"].mean(), linestyle="--"); #add horizantal average line
```

# Subplot of chol, age, thalach

```python
car_sales["Price"].plot();
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10))

scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# add data ax0
ax0.set(title="Heart Disease and Cholesterol levels", xlabel="Age", ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(), title="Target");

ax0.axhline(over_50["chol"].mean(), linestyle="--"); #add horizantal average line

scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"]);
# add data ax1
ax1.set(title="Heart Disease and Max heart rate", xlabel="Age", ylabel="Max Heart rate")
ax1.legend(*scatter.legend_elements(), title="Target");

ax1.axhline(over_50["thalach"].mean(), linestyle="--"); #add horizantal average line

#add title to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
